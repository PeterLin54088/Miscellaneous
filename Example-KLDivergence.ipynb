```python
import scipy
from scipy.stats import t
from scipy.stats import norm

import numpy as np

import matplotlib.pyplot as plt
```

```python
df = 2
mean, var, skew, kurt = t.stats(df, moments='mvsk')
```

```python
x = np.linspace(-30, 30, 500)
plt.plot(x, t.pdf(x, df),
       'r-', lw=5, alpha=0.6, label='t pdf')
plt.plot(x, norm.pdf(x),
       'b-', lw=5, alpha=0.6, label='norm pdf')
plt.xlim(-5,5)
```

```python
print(np.sum(scipy.stats.entropy(norm.pdf(x))))
print(np.sum(scipy.stats.entropy(t.pdf(x,df))))
print(np.sum(scipy.stats.entropy(norm.pdf(x), t.pdf(x, df))))
print(np.sum(scipy.stats.entropy(t.pdf(x, df), norm.pdf(x))))
```
