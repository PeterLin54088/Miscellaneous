```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_vertical_modes():
    def mode(z, m=1):
        """
        Compute the vertical normal mode for a given vertical mode number (m) under the assumption of constant stratification.

        Parameters:
        z : array-like
            Nondimensional height.
        m : int, optional
            Vertical mode number, default is 1.

        Returns:
        array-like
            The vertical normal mode, computed as a cosine function.
        """
        return np.cos(m * np.pi * z)

    # Generate nondimensional height values (z) from 0 to -1
    z = np.linspace(0, -1, 512)

    # Create a figure with appropriate size and resolution
    fig, ax = plt.subplots(figsize=(9, 16), dpi=120)

    # Ref
    ax.plot([0, 0], [-1, 0], "k--", lw=1.5)
    # Plot vertical normal modes for m = 1 to m = 4
    for m in range(1, 5):
        ax.plot(np.flip(mode(z, m)), z, label=f"m = {m}", lw=1.2)

    # Add legend and axis labels
    ax.legend()
    ax.set_xlabel("Amplitude", fontsize = 16)
    ax.set_ylabel("Nondimensional Height", fontsize = 16)
    ax.set_xticks(np.linspace(-1,1,5))
    ax.set_yticks(np.linspace(-1,0,6))
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 0)
    # title
    ax.set_title("Vertical Normal Modes", fontsize = 22)
    # Return the figure object
    return fig


# To use the function and show the plot:
fig = display_vertical_modes()
fig.show()
```
