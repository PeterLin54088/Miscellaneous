# Vertical Transform (Vertical Normal Mode)

---

## Introduction
In geophysical fluid dynamics, the governing equations often involve complex interactions between **horizontal and vertical components**. These equations can become particularly complicated when **vertical derivatives** are involved. However, in many practical situations, it is observed that the **vertical structure** of solutions is relatively simple, especially under hydrostatic approximation. To take advantage of this, we introduce the concept of **vertical transformation**, which allows us to convert the problem into a more manageable form.

---

## Derivation of the Simplified Equation Set
To facilitate the application of vertical transformation, we start by considering a set of simplified equations. These equations are derived under the following assumptions:
- **Hydrostatic approximation**
- **Boussinesq approximation**
- **Linearization around a state of rest (perturbation variables)**

**Fundamental Equations:**

- Horizontal Momentum Equations:
  $$
  \frac{\partial u}{\partial t} - fv = -\frac{\partial \phi}{\partial x}
  $$
  $$
  \frac{\partial v}{\partial t} + fu = -\frac{\partial \phi}{\partial y}
  $$

- Vertical Momentum Equation:
  $$
  0 = -\frac{\partial \phi}{\partial z} + b
  $$

- Continuity Equation:
  $$
  \nabla \cdot u + \frac{\partial \omega}{\partial z} = 0
  $$

- Thermodynamic Equation:
  $$
  \frac{\partial b}{\partial t} + \omega N^2 = 0
  $$

**Reference**: Vallis, Atmospheric and Oceanic Fluid Dynamics (pg.133)

> *The equations are more general than needed, as the vertical structure is often simple in linear problems.*

---

## The Motivation for Vertical Transformation
One of the key challenges in solving the governing equations is managing the vertical derivative operator. By combining all equations involving vertical derivatives, we arrive at a composite equation:
$$
\frac{\partial}{\partial t}\frac{\partial}{\partial z}\left(\frac{1}{N^2}\frac{\partial \phi}{\partial z}\right) - \nabla \cdot u=0
$$
The goal is to find a transformation that **simplifies the vertical derivative operator**, effectively reducing the problem from a 3D framework to a horizontal problem with straightforward vertical normal modes. This approach significantly reduces computational complexity and improves the clarity of physical interpretation.

---

## Eigenspace: From Derivation to Multiplication

**Recap: Transforming Complex Operators**

In mathematical analysis, one of the fundamental ideas is that a **complex operator in one space can become simpler in another space**. A classic example is the Fourier transform, which converts differentiation (a complex operation) into multiplication (a simpler operation):
$$\mathcal{F}{\{f^{\prime}(x)\}} = iw\mathcal{F}{\{f(x)\}}$$
This insight motivates us to find a similar transformation for the **vertical derivative operator** in geophysical fluid dynamics. We seek a space where the operator becomes **multiplicative**, greatly simplifying our problem.

**Eigenvalue Problem: Finding the Simplifying Space**

To achieve this, we look for eigenfunctions ($\Psi_m$) that satisfy the eigenvalue equation:
$$
\mathcal{L}(\Psi_m) = \lambda_m \Psi_m
$$
where the operator $\mathcal{L}$ is:
$$
\mathcal{L} = \frac{\partial}{\partial z} \left(\frac{1}{N^2} \frac{\partial }{\partial z}\right)
$$

**Separation of Variables**

To isolate the vertical component, we decompose the solution into horizontal and vertical parts:
$$
[u,v,\phi]=\sum^{\infty}_{m=0}{[u_m(x,y,t),v_m(x,y,t),\phi_m(x,y,t)] \cdot \Psi_m(z)}
$$

**Projection Operator and Reconstruction**

To project the original problem into the **eigenfunction space**, we define a **projection operator ($\mathcal{T}_m$)**:
$$
\mathcal{T}_m(f(z)) = \frac{1}{\|\Psi_m(z)\|} \langle f(z), \Psi_m(z) \rangle = \hat{f}_m
$$

The original function can then be re-expressed as a sum over eigenfunctions:
$$
f = \sum_{m} \mathcal{T}_m(f) \cdot \Psi_m
$$

---

## Derivation of the Transformed Equations

**Applying the Projection:**
Substitute the **mode expansion** into the composite equation:
$$
\frac{\partial }{\partial t} \mathcal{L} \left(\sum_{m} \mathcal{T}_m(\phi) \cdot \Psi_m \right) - \nabla \cdot \left(\sum_{m} \mathcal{T}_m(u) \cdot \Psi_m\right) = 0
$$

**Interchanging Summation and Operator:**
$$
\sum_{m} \left( \frac{\partial \mathcal{T}_m(\phi)}{\partial t} \cdot \mathcal{L}(\Psi_m) \right) - \sum_{m} \left((\nabla \cdot \mathcal{T}_m(u)) \cdot \Psi_m\right) = 0
$$

**Using Eigenfunction Properties:**
$$
\frac{\partial \hat{\phi}_m}{\partial t} - \frac{1}{\lambda_m} \nabla \cdot \hat{u}_m = 0
$$

---

## Transformed Equation Set: Vertical Normal Modes

After applying the transformation, the equations for each mode become:
$$
\frac{\partial \hat{u}_m}{\partial t} - f \hat{v}_m = -\frac{\partial \hat{\phi}_m}{\partial x}
$$
$$
\frac{\partial \hat{v}_m}{\partial t} + f \hat{u}_m = -\frac{\partial \hat{\phi}_m}{\partial y}
$$
$$
\frac{\partial \hat{\phi}_m}{\partial t} - \frac{1}{\lambda_m} \nabla \cdot \hat{u}_m = 0
$$

---

## How to Determine Eigenfunctions and Eigenvalues?

To find the eigenfunctions and eigenvalues, we follow the methodology outlined by Scott R. Fulton and Wayne H. Schubert (1985) and Vallis (pg.135). The problem can be approached in two scenarios:
- Constant $N^2$
- Non-constant $N^2$

In this note, we focus on the analytical solution for the constant $N^2$ case. While the non-constant $N^2$ scenario can also be addressed, it typically requires numerical methods and is beyond our current scope.

**Constant $N^2$**

Given that $N^2$ is a constant, the problem becomes straightforward --- we utilize the Fourier transform to project the functions onto Fourier bases. Then, applying Neumann boundary conditions allows us to precisely determine the coefficients.

**Governing Equation**

The governing differential equation in this context is given by:
$$\mathcal{L}(\Psi_m) = \frac{1}{N^2} \frac{\partial^2 \Psi_m}{\partial z^2} = \lambda_m \Psi_m$$

The general solution to this equation is:
$$\Psi_m = A e^{\sqrt{\lambda_m N^2} z} + B e^{-\sqrt{\lambda_m N^2} z}$$

**Boundary Conditions**

From Scott R. Fulton and Wayne H. Schubert (1985), the boundary conditions are:
$$\frac{\partial \Psi_m}{\partial z}(0)=\frac{\partial \Psi_m}{\partial z}(-H)=0$$
Applying boundary conditions yield:
1. At $z = 0$:
$$\frac{\partial \Psi_m}{\partial z}(0) = 0 \Longrightarrow A - B = 0 \Longrightarrow \Psi_m = \cos{(\sqrt{\lambda_m N^2} z)}$$
2. At $z = -H$:
$$\frac{\partial \Psi_m}{\partial z}(-H) = 0 \Longrightarrow \Psi_m = \cos{(\sqrt{\lambda_m N^2} H)} = 0$$
For the above expression to hold, the argument of the cosine function must satisfy:
$$\sqrt{\lambda_m N^2} H = m\pi \space, \forall{m} \in Z$$

**Eigenvalues and Eigenfunctions**

From the boundary condition analysis, we obtain the eigenvalue as:
$$\lambda_m = \left(\frac{m \pi}{N H}\right)^2$$
The corresponding eigenfunction is:
$$\Psi_m = \cos{(m \pi\frac{z}{H})}$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def vertical_normal_mode_analytical(z: np.ndarray, m: int = 1):
    """
    Compute the analytical solution for the vertical normal mode eigenfunctions.

    Parameters:
    -----------
    z : np.ndarray
        1D array of vertical coordinates (typically normalized between 0 and 1).
    m : int, optional
        Mode number (default is 1).

    Returns:
    --------
    np.ndarray
        Eigenfunction values corresponding to the vertical mode `m`.
    """
    eigenfunctions = np.cos(m * np.pi * z)
    return eigenfunctions


def vertical_normal_mode_FDM(N: int, m: int = 1):
    """
    Approximate the vertical normal mode eigenfunction using the finite difference method (FDM).

    This function constructs a second-order central finite difference matrix with Neumann 
    boundary conditions at both ends, as described in:
    https://www.math.toronto.edu/mpugh/Teaching/Mat1062/notes2.pdf

    Parameters:
    -----------
    N : int
        Number of vertical grid points.
    m : int, optional
        Mode number to extract (default is 1).

    Returns:
    --------
    np.ndarray
        The m-th vertical mode eigenfunction (normalized such that its maximum absolute value is 1).
    """
    operator = (
        -2 * np.eye(N) + np.roll(np.eye(N), 1, axis=0) + np.roll(np.eye(N), -1, axis=0)
    )
    # Apply Neumann boundary conditions
    operator[0, 0] = -2
    operator[0, 1] = 2
    operator[0, -1] = 0
    operator[-1, 0] = 0
    operator[-1, -2] = 2
    operator[-1, -1] = -2

    eigenvalues, eigenvectors = np.linalg.eig(operator)
    eigenvectors = eigenvectors[:, np.argsort((np.abs(eigenvalues)))]
    mode = eigenvectors[:, m]

    # Ensure consistent sign (positive at the surface)
    if mode[0] < 0:
        mode = -mode

    # Normalize
    return mode / np.max(np.abs(mode))
```

```python
def display_vertical_modes(method="analytical"):
    """
    Plot the first four vertical normal modes using either the analytical solution
    or the finite difference approximation.

    Parameters:
    -----------
    method : str, optional
        Method used to compute vertical modes. Options are:
        - "analytical": uses the analytical cosine-based solution.
        - "finite_difference": uses finite difference matrix approximation.

    Returns:
    --------
    matplotlib.figure.Figure
        The figure object containing subplots of the first four vertical modes.

    Raises:
    -------
    ValueError
        If an invalid method is specified.
    """
    z = np.linspace(0, -1, 128)
    fig, axes = plt.subplots(
        nrows=2, ncols=2, sharex=True, sharey=True, figsize=(9, 16), dpi=160
    )

    for (i, j), ax in np.ndenumerate(axes):
        m = i * 2 + j + 1
        if method == "analytical":
            vertical_mode = vertical_normal_mode_analytical(z, m)
        elif method == "finite_difference":
            vertical_mode = vertical_normal_mode_FDM(len(z), m)
        else:
            raise ValueError(
                "Invalid method. Choose 'analytical' or 'finite_difference'."
            )
        ax.plot([0, 0], [-1, 0], "k--", lw=1.2)
        ax.plot(vertical_mode, z, "k-", lw=1.2)
        ax.fill_betweenx(z, vertical_mode)
        ax.set_title(f"m = {m}")
        ax.set_xticks(np.linspace(-1, 1, 3))
        ax.set_yticks(np.linspace(-1, 0, 3))
        ax.set_xlim(-1, 1)
        ax.set_ylim(-1, 0)

    fig.suptitle("Vertical Normal Mode", fontsize=16)
    fig.tight_layout()
    return fig
```

```python
fig = display_vertical_modes(method="analytical")
fig.show()
```

```python
fig = display_vertical_modes(method="finite_difference")
fig.show()
```
