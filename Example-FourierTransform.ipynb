## **Brief Summary for Continuous Fourier Transform**

### **Basic Properties**

The Fourier Transform possesses several fundamental properties that facilitate signal analysis and manipulation. Below are some key properties:

- **Linearity**  
  The Fourier Transform of a linear combination of functions follows the same linearity principle:
  $$
  F(ax_1(t) + bx_2(t)) = aF(x_1(t)) + bF(x_2(t)) = aX_1(\omega) + bX_2(\omega)
  $$

- **Scaling**  
  Scaling in the time domain affects both the amplitude and frequency in the Fourier domain:
  $$
  x(at) \longleftrightarrow \frac{1}{|a|}X\left(\frac{\omega}{a}\right)
  $$

- **Time Shifting**  
  A time shift in the original function introduces a phase shift in the frequency domain:
  $$
  F(x(t - t_0)) = e^{-i \omega t_0} F(x(t))
  $$

- **Duality**  
  The Fourier Transform of a function and its inverse relationship follow the duality principle:
  $$
  x(t) \longleftrightarrow X(\omega)
  $$
  $$
  X(t) \longleftrightarrow 2\pi \cdot x(-\omega)
  $$

> **Note:**  
> The factor $2\pi$ in the duality property depends on the choice of normalization.

### **Advanced Properties**

Beyond the basic properties, Fourier Transform exhibits advanced properties that are crucial for signal processing:

- **Multiplication (Modulation Property)**  
  The multiplication of two functions in the time domain corresponds to their convolution in the frequency domain:
  $$
  p(t) \cdot q(t) \longleftrightarrow P(f) * Q(f)
  $$

- **Convolution**  
  The convolution of two signals in the time domain is equivalent to their multiplication in the frequency domain:
  $$
  p(t) * q(t) \longleftrightarrow P(f) \cdot Q(f)
  $$

- **Frequency Shifting**  
  Shifting a function in the frequency domain corresponds to modulation in the time domain:
  $$
  X(\omega - \omega_0) \longleftrightarrow x(t)e^{i \omega_0 t}
  $$

Here, $*$ represents the convolution operator, while $\cdot$ denotes multiplication.

## **From CFT to DFT: A Step-by-Step Understanding**

This section outlines the conceptual progression from the **Continuous-Time Fourier Transform (CFT)** to the **Discrete Fourier Transform (DFT)**.

- Part 1: We begin by examining how **sampling** and **windowing** affect signals in the CFT domain. This sets the stage for understanding key spectral phenomena such as **aliasing** and **spectral leakage**.

- Part 2: Next, we apply sampling to continuous-time signals to derive the **Discrete-Time Fourier Transform (DTFT)**, which characterizes the frequency content of signals defined over *infinite* discrete time.

- Part 3: We then introduce windowing to truncate the signal in time, which yields a *finite-length* discrete-time signal. Although the signal is finite in time, its DTFT remains a continuous and periodic function of frequency.

- Part 4: Finally, by **sampling the DTFT** in the frequency domain, we obtain the DFT — a fully discrete, finite-dimensional representation of the signal’s spectrum, suitable for numerical computation and practical implementation.

---

## **Part 1: Effects of Sampling and Windowing in CFT**

We begin in the continuous-time domain by examining how sampling and windowing influence a signal’s frequency-domain representation.

### 1.1 Sampling in CFT

Sampling can be modeled using a Dirac comb:

$$
s(t) = \sum_{n = -\infty}^{\infty} \delta(t - nT)
$$

Fourier Transform of the Dirac comb:

$$
\hat{s}(\omega) = \frac{1}{T} \sum_{n = -\infty}^{\infty} \delta\left(\omega - \frac{n}{T} \right)
$$

Sampling a signal $f(t)$:

$$
f_s(t) = f(t) \cdot s(t)
$$

In the frequency domain, this results in:

$$
\hat{f}_s(\omega) = \hat{f}(\omega) * \hat{s}(\omega)
= \frac{1}{T} \sum_{n = -\infty}^{\infty} \hat{f}\left( \omega - \frac{n}{T} \right)
$$

**Key Insight**: Sampling causes periodic replication of $\hat{f}(\omega)$ at intervals of $\frac{1}{T}$ in the frequency domain. If the repetitions overlap, different frequency components become indistinguishable — a phenomenon known as **aliasing**.

---

### 1.2 Windowing in CFT

Use a rectangular window (symmetric, center at $t=0$):

$$
w(t) = \begin{cases}
1 & |t| \leq \frac{L}{2} \\
0 & \text{otherwise}
\end{cases}
$$

Fourier Transform of the window:

$$
\hat{w}(\omega) = L \cdot \mathrm{sinc}(\omega L)
$$

Windowed signal:

$$
f_w(t) = f(t) \cdot w(t)
$$

In the frequency domain:

$$
\hat{f}_w(\omega) = \hat{f}(\omega) * \hat{w}(\omega)
= \hat{f}(\omega) * \left[L \cdot \mathrm{sinc}(\omega L)\right]
$$

**Key Insight**: Windowing acts like a moving average applied to the frequency spectrum — it smooths out sharp features, causing the original spectrum to blur and spread. As a result, energy from one frequency "leaks" into nearby frequencies, a phenomenon known as **spectral leakage**.

---

## **Part 2: Bridging the Gap — From CFT to DTFT**

By following "Section 1.1, Sampling in CFT", and re-expressing the signal using a discrete-time representation, we obtain the **Discrete-Time Fourier Transform (DTFT)**.

### 2.1 Sampling a Continuous-Time Signal

Following the same procedure as in 1.1, applying sampling:

$$
f_s(t) = f(t) \cdot \sum_{n = -\infty}^{\infty} \delta(t - nT)
= \sum_{n = -\infty}^{\infty} f(nT) \cdot \delta(t - nT)
$$

Taking the Fourier Transform:

$$
\hat{f}_s(\omega) = \sum_{n = -\infty}^{\infty} f(nT) \cdot e^{-i \omega n T}
$$

Here, we re-express the *continuous-time* signal as a *discrete-time* signal, $f[n] = f(nT)$:

$$
\hat{f}_s(\omega) = \sum_{n = -\infty}^{\infty} f[n] \cdot e^{-i \omega n T}
$$

This is the **Discrete-Time Fourier Transform (DTFT)**.

---


### 2.2 DTFT vs. CFT

Taking the Fourier Transform of the sampled signal yields a frequency-domain expression involving the discrete-time sequence $f[n] = f(nT)$. This is recognized as the Discrete-Time Fourier Transform (DTFT). In short,

> **CTFT + Sampling → DTFT.**

- **CFT:**
  $$
  \mathcal{F}_{CFT}{\left\{ f(t) \right\}} = \int_{-\infty}^{\infty} f(t) \cdot e^{-i \omega t} dt
  $$
  
- **DTFT:**
  $$
  \mathcal{F}_{DTFT}{\left\{ f[n] \right\}} = \sum_{n = -\infty}^{\infty} f[n] \cdot e^{-i \omega n T}
  $$

| Feature     | CFT                       | DTFT                          |
| ----------- | ------------------------- | ----------------------------- |
| Time domain | Continuous                | Discrete                      |
| Frequency   | Continuous (non-periodic) | Continuous (but **periodic**) |
| Integration | Infinite integral         | Infinite sum                  |

---

## **Part 3: Practical Considerations — Windowing and Finite DTFT**

To make the DTFT *numerically* computable, we apply windowing to truncate the infinite-length discrete signal, yielding a **finite-length discrete-time signal**.

### 3.1 Sampling & Windowing a Continuous-Time Signal

Apply both sampling and windowing to a continuous-time signal:

$$
f_{sw}(t) = f(t) \cdot w(t) \cdot \sum_{n=-\infty}^{\infty} \delta(t - nT), \quad
w(t) = \begin{cases}
1 & |t| \leq \frac{L}{2} \\
0 & \text{otherwise}
\end{cases}
$$

Sampling the result yields the discrete signal:

$$
f_{sw}[n] = f(nT) \cdot w(nT) = f[n] \cdot w[n]
$$

---

### 3.2 DTFT of a Windowed, Sampled Signal

We compute the DTFT of the sampled and windowed signal:

$$
\hat{f}_{sw}(\omega) = \sum_{n = -\infty}^{\infty} f[n] \cdot w[n] \cdot e^{-i \omega nT}
$$

Since $w[n] = 0$ when $|nT| > \frac{L}{2}$, the sum becomes finite:

$$
\hat{f}_{sw}(\omega) = \sum_{n = n_{min}}^{n_{max}} f[n] \cdot e^{-i \omega nT}, \quad
n_{min} = \left\lfloor -\frac{L}{2T} \right\rfloor, \quad n_{max} = \left\lfloor \frac{L}{2T} \right\rfloor
$$

**Key Insight**: Windowing allows us to compute the DTFT of a finite-length signal. However, this introduces spectral leakage due to convolution with the sinc-like function.

---

## **Part 4: From DTFT to DFT — Toward Implementation**

To implement the DTFT in practice, we sample it at discrete frequency points to obtain the Discrete Fourier Transform (DFT). Before we can do this, we make two key adjustments:

1. Use a causal window (starting at $n = 0$)
2. Normalize the frequency variable

### 4.1 Causal Windowing

In practical applications, signals are often indexed starting at $n = 0$, rather than being centered around zero. To reflect this, we use a *causal window*, defined as:

$$
w(t) = \begin{cases}
1 & 0 \leq t \leq L \\
0 & \text{otherwise}
\end{cases}
$$

This window is called causal because it starts at $t = 0$ and extends forward in time, as is typical for real-world data acquisition.

Applying the same sampling and windowing procedure from **Part 3**, but with this causal window, we obtain a shifted DTFT expression:


$$
\hat{f}_{sw}(\omega) = \sum_{n = 0}^{N-1} f[n] \cdot e^{-i \omega nT}, \quad
N = \frac{L}{T}
$$

---

### 4.2 Re-expressing the DTFT Expression

We re-express the DTFT to prepare for sampling:

$$
\hat{f}_{sw}(\omega) = \sum_{n = 0}^{N-1} f[n] \cdot e^{-i \omega \frac{n}{N}L}
$$

Let $\omega^{*} = \omega L$ (normalized frequency):

$$
\hat{f}_{sw}(\omega^{*}) = \sum_{n = 0}^{N-1} f[n] \cdot e^{-i \omega^{*} \frac{n}{N}}
$$

> Why does numpy.fft.fftfreq() require sample spacing $d$? Because DFT frequency bins are normalized, and $d$ lets us scale back to real-world frequency units!
---

### 4.3 Sampling the DTFT — Generalized DFT

Since $\hat{f}_{sw}(\omega^*)$ is $2\pi$-periodic in $\omega^*$, we only need to evaluate it over a fundamental interval — typically $\omega^* \in [-\pi, \pi)$ or $\omega^* \in [0, 2\pi)$.

To obtain a discrete, finite representation of the frequency content, we sample the DTFT at $M$ equally spaced points within $[0, 2\pi)$:

$$
\omega^{*}_k = \frac{2\pi k}{M}, \quad k = 0, 1, \dots, M-1
$$

This yields the **Generalized DFT**:

$$
\hat{f}_{sw}[k] = \sum_{n = 0}^{N-1} f[n] \cdot e^{-i \omega^{*}_k \frac{n}{N}}
$$

---

### 4.4 Information-Theoretic Viewpoint — From Generalized DFT to Standard DFT

> A signal of length $N$ contains $N$ pieces of information.

- $M < N$: under-sampling → information loss
- $M > N$: over-sampling → interpolation only
- $M = N$: balanced → lossless and invertible

This gives us the **Standard DFT**:

$$
\omega^{*}_k = \frac{2\pi k}{N}, \quad k = 0, 1, \dots, N-1
$$


$$
\hat{f}_{sw}[k] = \sum_{n = 0}^{N-1} f[n] \cdot e^{-i \omega^{*}_k \frac{n}{N}}
$$



## **Discrete Fourier Transform**

### **Definition of DFT and iDFT**
The Discrete Fourier Transform (DFT) and its inverse (iDFT) are defined as follows:

$$
f_n = \frac{1}{\sqrt{N}} \sum_{k=0}^{N-1}{F_k \cdot e^{i 2\pi k \frac{n}{N}}}, \quad n \in [0, N-1]
$$

$$
F_k = \frac{1}{\sqrt{N}} \sum_{n=0}^{N-1}{f_n \cdot e^{-i 2\pi k \frac{n}{N}}}, \quad k \in [0, N-1]
$$

where:

- $F_k$ represents the frequency-domain coefficients.
- $f_n$ denotes the time-domain (or space-domain) signal.

> Note: This is a unitary definition of the DFT, which differs from many software implementations like numpy.fft.fft()

### **Matrix Representation of DFT and iDFT**

For numerical implementation, we expressed DFT and iDFT in matrix form.

We define the DFT matrix $\mathbf{W}$ and the iDFT matrix $\mathbf{W^{-1}}$ as:

$$
\mathbf{W}_{nk} = \frac{1}{\sqrt{N}} e^{-i 2\pi k \frac{n}{N}}
$$

$$
\mathbf{W^{-1}}_{nk} = \mathbf{\overline{W}}_{nk} = \frac{1}{\sqrt{N}} e^{i 2\pi k \frac{n}{N}}
$$

Using this notation, the forward and inverse DFT can be represented as matrix multiplications:

$$
\mathbf{f} = \mathbf{W^{-1}} \mathbf{F}
$$

$$
\mathbf{F} = \mathbf{W} \mathbf{f}
$$

**Structure of the DFT and iDFT Matrices**

The elements of the matrices $\mathbf{W}$ and $\mathbf{W^T}$ are as follows:

> **DFT Matrix Representation:**
> $$
> \mathbf{W} = \frac{1}{\sqrt{N}}
> \begin{bmatrix}
> e^{-i 2\pi \cdot 0 \cdot \frac{0}{N}} & e^{-i 2\pi \cdot 0 \cdot \frac{1}{N}} & \dots & e^{-i 2\pi \cdot 0 \cdot \frac{N-1}{N}} \\
> e^{-i 2\pi \cdot 1 \cdot \frac{0}{N}} & e^{-i 2\pi \cdot 1 \cdot \frac{1}{N}} & \dots & e^{-i 2\pi \cdot 1 \cdot \frac{N-1}{N}} \\
> \vdots & \vdots & \ddots & \vdots \\
> e^{-i 2\pi \cdot (N-1) \cdot \frac{0}{N}} & e^{-i 2\pi \cdot (N-1) \cdot \frac{1}{N}} & \dots & e^{-i 2\pi \cdot (N-1) \cdot \frac{N-1}{N}}
> \end{bmatrix}
> $$

> **Inverse DFT (iDFT) Matrix Representation:**
> $$
> \mathbf{W^{-1}} = \frac{1}{\sqrt{N}}
> \begin{bmatrix}
> e^{i 2\pi \cdot 0 \cdot \frac{0}{N}} & e^{i 2\pi \cdot 0 \cdot \frac{1}{N}} & \dots & e^{i 2\pi \cdot 0 \cdot \frac{N-1}{N}} \\
> e^{i 2\pi \cdot 1 \cdot \frac{0}{N}} & e^{i 2\pi \cdot 1 \cdot \frac{1}{N}} & \dots & e^{i 2\pi \cdot 1 \cdot \frac{N-1}{N}} \\
> \vdots & \vdots & \ddots & \vdots \\
> e^{i 2\pi \cdot (N-1) \cdot \frac{0}{N}} & e^{i 2\pi \cdot (N-1) \cdot \frac{1}{N}} & \dots & e^{i 2\pi \cdot (N-1) \cdot \frac{N-1}{N}}
> \end{bmatrix}
> $$

```python
def fftfreq(n: int, d: int, shift: bool = False) -> np.ndarray:
    """
    Compute the discrete Fourier Transform sample frequencies.

    Parameters:
        n (int): Number of points in the Fourier transform.
        d (int): Sample spacing (inverse of tfe sampling rate).
        shift (bool, optional): Whether to shift the zero frequency to the center. Defaults to False.

    Returns:
        np.ndarray: Array of frequency values.
    """
    modes = np.arange(np.ceil(-n / 2), np.ceil(n / 2), 1)
    total_spacing = n * d

    if not shift:
        modes = np.roll(modes, np.ceil(len(modes) / 2))

    wavenumber = modes / total_spacing
    return wavenumber


def fftshift(signal: np.ndarray) -> np.ndarray:
    """
    Shift the zero-frequency component to the center of the spectrum.

    Parameters:
        signal (np.ndarray): Input array representing the Fourier-transformed data.

    Returns:
        np.ndarray: Shifted array with zero frequency centered.
    """
    shift = tuple(int(np.floor(dim / 2)) for dim in np.shape(signal))
    axis = tuple(range(signal.ndim))

    return np.roll(signal, shift, axis)


def ifftshift(signal: np.ndarray) -> np.ndarray:
    """
    Inverse shift of the zero-frequency component to its original position.

    Parameters:
        signal (np.ndarray): Input array that has been shifted by `fftshift`.

    Returns:
        np.ndarray: Array with zero frequency moved back to its original location.
    """
    shift = tuple(-int(np.floor(dim / 2)) for dim in np.shape(signal))
    axis = tuple(range(signal.ndim))
    return np.roll(signal, shift, axis)
```

```python
def DFT(signal: np.ndarray) -> np.ndarray:
    """
    Compute the Discrete Fourier Transform (DFT) of a given signal.

    Args:
        signal (np.ndarray): Input signal, a 1D array of complex or real values.

    Returns:
        np.ndarray: The transformed signal in the frequency domain.
    """
    N = len(signal)
    sample_indices = np.arange(N) / N

    wavenumbers = np.arange(N).reshape((N, 1))

    bases = np.exp(-1j * 2 * np.pi * wavenumbers * sample_indices)

    return bases @ signal / np.sqrt(N)


def iDFT(spectrum: np.ndarray) -> np.ndarray:
    """
    Compute the Inverse Discrete Fourier Transform (iDFT) of a given frequency spectrum.

    Args:
        spectrum (np.ndarray): Input spectrum, a 1D array of complex frequency components.

    Returns:
        np.ndarray: The reconstructed time-domain signal.
    """
    N = len(spectrum)
    sample_indices = np.arange(N) / N

    wavenumbers = np.arange(N).reshape((N, 1))

    bases = np.exp(1j * 2 * np.pi * wavenumbers * sample_indices)

    return bases @ spectrum / np.sqrt(N)
```

```python
class BasicTest_FourierAnalysis(unittest.TestCase):
    def test_fftfreq(self):
        for N in np.random.randint(128, 256, size=100):
            with self.subTest(N=N):
                self.assertTrue(
                    np.allclose(fftfreq(N, 1 / N), np.fft.fftfreq(N, 1 / N)),
                    f"fftfreq failed for N={N}",
                )

    def test_fftshift(self):
        for N in np.random.randint(128, 256, size=100):
            signal = np.arange(N)
            with self.subTest(N=N):
                self.assertTrue(
                    np.allclose(fftshift(signal), np.fft.fftshift(signal)),
                    f"fftshift failed for N={N}",
                )
                self.assertTrue(
                    np.allclose(ifftshift(signal), np.fft.ifftshift(signal)),
                    f"ifftshift failed for N={N}",
                )

    def test_DFT_iDFT(self):
        for N in np.random.randint(128, 256, size=100):
            time = np.linspace(0, 1 + np.random.rand(), N, endpoint=False)
            signal = np.exp(-1j * 2 * np.pi * 0 * time).real
            for wavenumber in np.random.randint(1, 10, size=2):
                signal += (1 / wavenumber) * np.exp(
                    -1j * 2 * np.pi * wavenumber * time
                ).real
            signal += np.random.normal(loc=0, scale=0.5, size=np.shape(signal))
            with self.subTest(N=N):
                self.assertTrue(
                    np.allclose(np.fft.fft(signal, norm="ortho"), DFT(signal)),
                    f"DFT failed for N={N}",
                )
                self.assertTrue(
                    np.allclose(np.fft.ifft(signal, norm="ortho"), iDFT(signal)),
                    f"iDFT failed for N={N}",
                )
```

## **Power Spectrum Density and Phase Angle**

The power spectral density describes how the power of a signal is distributed over frequency. It can be defined in both continuous and discrete contexts.

- **CFT**:
$$S_{ff}{(\omega)} = \lim_{T \rightarrow \infty}{\frac{1}{T}\left\vert \hat{f}(\omega)\right\vert^2}$$
- **DFT**:
$$S_{ff}[k] = \frac{1}{N}\left\vert \hat{f}[k]\right\vert^2$$

The phase spectrum characterizes the phase shift of each frequency component of the signal. It is computed directly from the complex-valued Fourier coefficients:

- **CFT**:
$$\phi{(\omega)} = \mathrm{arctan2}{\left( \operatorname{Im}{(\hat{f}{(\omega)})}, \operatorname{Re}{(\hat{f}{(\omega)})} \right)}$$
- **DFT**:
$$\phi[k] = \mathrm{arctan2}{\left( \operatorname{Im}{(\hat{f}{[k]})}, \operatorname{Re}{(\hat{f}{[k]})} \right)}$$

This returns the angle (in radians) of the complex number $\hat{f}{(\omega)}$, capturing the relative timing (phase) of each frequency component.


## **Parsevel's Identity**

Parseval’s identity ensures that the total energy (or average power) of a signal is preserved between the time and frequency domains.

- **CFT**:
$$P = \int_{-\infty}^{\infty}{\left[ \lim_{T \rightarrow \infty}{\frac{1}{T}\left\vert f(t)\right\vert^2}\right] dt} = \int_{-\infty}^{\infty}{\left[ \lim_{T \rightarrow \infty}{\frac{1}{T}\left\vert \hat{f}(\omega)\right\vert^2}\right] d\omega} $$
- **DFT**:
$$P = \frac{1}{N}\sum_{n = 0}^{N-1}{\left\vert f[n]\right\vert^2} = \frac{1}{N}\sum_{\omega = 0}^{N-1}{\left\vert \hat{f}[k]\right\vert^2}$$

This states that the average power of a discrete signal is equal in both the time and frequency domains.

> **Note**: The Fourier Transform used above is assumed to be a **unitary transform**, i.e., using `norm="ortho"` in FFT implementations. This ensures energy is preserved without additional scaling.

```python
def display_power_phase_spectrum():
    """
    Synthesizes a complex-valued, unit-norm signal using a set of randomly selected
    significant frequency components, then visualizes its power spectrum and phase spectrum.
    Also verifies Parseval's identity under the assumption of a unitary Fourier transform.

    Workflow:
        1. Randomly selects `n` significant frequency bins within a limited bandwidth ±f_0.
        2. Constructs a signal using full-spectrum synthesis:
            - Significant frequencies contribute with equal amplitude and fixed phase shift.
            - All other frequencies are assigned a very small magnitude to minimize numerical artifacts.
        3. Computes the Power Spectral Density (PSD) and Phase Spectrum via unitary FFT.
        4. Verifies Parseval’s identity by comparing time-domain and frequency-domain average power.
        5. Plots:
            - Power Spectrum (magnitude squared of FFT coefficients)
            - Phase Spectrum (angle of FFT coefficients in degrees)

    Notes:
        - Time is normalized and unitless over [0, 1), so frequency units are also normalized (cycles per unit interval).
        - FFT is computed with `norm="ortho"` to ensure energy preservation (unitary transform).

    Returns:
        None. The function displays plots and prints whether Parseval’s identity holds.
    """

    def frequencies_bins_picker(
        time: np.ndarray, f_0: float = 15.0, n: int = 3
    ) -> float:
        """
        Selects and returns a sorted array of `n` random frequency bins within ±f_0.

        Parameters:
            time (np.ndarray): time over [0, 1).
            f_0 (float, optional): Cutoff for selecting frequency bins.
            n (int, optional): Number of frequency bins to randomly select.

        Returns:
            np.ndarray: Sorted array of `n` randomly selected frequency bins within ±f_0.
        """
        frequencies = np.fft.fftfreq(time.size, d=1 / time.size)
        band_limited = frequencies[np.abs(frequencies) <= f_0]
        chosen = np.random.choice(band_limited, size=n, replace=False)
        chosen_sorted = np.sort(chosen)
        return chosen_sorted

    def signal_generator(time: np.ndarray, significant_freq: np.ndarray) -> np.ndarray:
        """
        Generates a complex-valued unit-norm signal.
        Significant frequencies contribute equally with unit energy, while non-significant
        bins are softly populated with small values to prevent numerical artifacts.

        Parameters:
            time (np.ndarray): time over [0, 1).
            significant_freq (np.ndarray): Array of frequency bins with dominant energy.

        Returns:
            np.ndarray: Complex-valued signal, constructed using full spectrum.
        """
        N = len(time)
        signal = np.zeros_like(time, dtype=complex)
        for freq in np.fft.fftfreq(N, d=1 / N):
            magnitude = (
                np.sqrt(1 / N) if freq in significant_freq else 1e-6 * np.sqrt(1 / N)
            )
            phase = 0.5 * np.pi if freq in significant_freq else 0.0
            signal += magnitude * np.exp(1j * (2 * np.pi * freq * time + phase))
        return signal

    def power_spectral_density(signa: np.ndarray) -> np.ndarray:
        """
        Computes the Power Spectral Density (PSD) of a signal using the Discrete Fourier Transform (DFT).
        The PSD represents the distribution of signal power across frequency components.

        Parameters:
            signal (np.ndarray): Complex or real-valued time-domain signal (1D array).

        Returns:
            np.ndarray: Power spectral density, computed as the squared magnitude of the
                        unitary DFT (i.e., FFT with norm="ortho").
        """
        return np.abs(np.fft.fft(signal, norm="ortho")) ** 2

    def phase_spectral_density(signal: np.ndarray) -> np.ndarray:
        """
        Computes the phase spectrum of a signal from its Discrete Fourier Transform (DFT).
        The phase spectrum indicates the phase angle (in degrees) of each frequency component.

        Parameters:
            signal (np.ndarray): Complex or real-valued time-domain signal (1D array).

        Returns:
            np.ndarray: Phase spectrum (in degrees), calculated using the arctangent of
                        the imaginary and real parts of the FFT.
        """
        return np.rad2deg(np.angle(np.fft.fft(signal)))

    time = np.linspace(0, 1, 200, endpoint=False)
    significant_freq = frequencies_bins_picker(time)
    signal = signal_generator(time, significant_freq)

    frequencies = np.fft.fftfreq(time.size, d=1 / time.size)
    power_spectrum = power_spectral_density(signal)
    phase_spectrum = phase_spectral_density(signal)

    time_domain_energy = np.mean(np.abs(signal) ** 2)
    freq_domain_energy = np.mean(power_spectrum)

    print(
        "Does Parseval's identity hold?",
        np.allclose(time_domain_energy, freq_domain_energy),
    )

    bar_width = (frequencies[1] - frequencies[0]) * 0.5
    fig, axes = plt.subplots(1, 2, figsize=(16, 9), dpi=160, sharex=True)

    axes[0].bar(frequencies, power_spectrum, width=bar_width, zorder=2)
    axes[0].grid(zorder=1)
    axes[0].set_title("Power Spectrum")
    axes[0].set_xlabel("Frequency")
    axes[0].set_ylabel("Power")

    axes[1].bar(frequencies, phase_spectrum, width=bar_width, zorder=2)
    axes[1].grid(zorder=1)
    axes[1].plot([-20, 20], [90, 90], "r--")
    axes[1].plot([0, 0], [-180, 180], "k-", lw=3, zorder=-1)
    axes[1].plot([-20, 20], [0, 0], "k-", lw=3, zorder=5)
    axes[1].set_title("Phase Spectrum")
    axes[1].set_xlabel("Frequency")
    axes[1].set_ylabel("Phase Angle (degrees)")
    axes[1].set_yticks(np.linspace(-180, 180, 9))
    axes[1].set_ylim(-180, 180)
    axes[1].set_xlim(-20, 20)

    fig.tight_layout()
    plt.show()

    return None
```

## **Aliasing**

### Recap

As derived in the **From CFT to DFT: A Step-by-Step Understanding** section, the sampling interval $\frac{1}{T}$ plays a crucial role in determining the effective bandwidth of the Discrete-Time Fourier Transform (DTFT). Specifically, the DTFT of a sampled signal can be expressed as:

$$
\hat{f}_s(\omega) = \hat{f}(\omega) * \hat{s}(\omega)
= \frac{1}{T} \sum_{n = -\infty}^{\infty} \hat{f}\left( \omega - \frac{n}{T} \right)
$$

This expression shows that the DTFT consists of repeated (periodic) copies of the original continuous Fourier transform $\hat{f}(\omega)$, spaced at interval $\frac{1}{T}$.

### Aliasing in DTFT 

If the original signal contains frequency components beyond half the sampling rate — the so-called Nyquist frequency — these shifted copies will overlap, resulting in aliasing. In this case, high-frequency components are indistinguishably folded into lower-frequency ones, distorting the spectral representation.

**Example**

Suppose a continuous-time signal has the following frequency-domain representation:

$$
\begin{cases}
\hat{f}(\omega_0) \neq 0 \quad &\text{(within Nyquist band)} \\
\hat{f}(\omega_0 + \frac{1}{T}) \neq 0 \quad &\text{(outside Nyquist band)} \\
\hat{f}(\omega) = 0 \quad &\text{for all other } \omega
\end{cases}
$$

Assume $\omega_0 \in \left(-\frac{1}{2T}, \frac{1}{2T} \right)$, so it lies within the DTFT’s effective bandwidth, while $\omega_0 + \frac{1}{T} \notin \left(-\frac{1}{2T}, \frac{1}{2T} \right)$, meaning it's beyond what the DTFT can resolve uniquely.

Then, after sampling, the DTFT is given by:

$$
\hat{f}_s(\omega) = \frac{1}{T} \sum_{n=-\infty}^{\infty} \hat{f}\left( \omega - \frac{n}{T} \right)
$$

Evaluating it at $\omega_0$ and $\omega_0 + \frac{1}{T}$, we get:

$$
\hat{f}_s(\omega_0) = \hat{f}_s(\omega_0 + \frac{1}{T}) = \frac{1}{T} \left( \hat{f}(\omega_0) + \hat{f}(\omega_0 + \frac{1}{T})\right)
$$

This shows that even though $\omega_0$ and $\omega_0 + \frac{1}{T}$ were distinct in the original signal, they now **alias** into the same frequency in the DTFT — they are **summed together**, and cannot be separated anymore. The DTFT has "folded" the high-frequency component into the lower one, due to its limited bandwidth.

### Aliasing in DFT

When we move from DTFT to DFT — a finite and discrete approximation — this aliasing is still present. The DFT samples the DTFT at equally spaced frequency bins, and any aliasing introduced by undersampling in the DTFT will be inherited and visible in the DFT as well.

```python
def display_aliasing():
    """
    Visualizes the effect of aliasing by comparing the power spectra of a signal
    sampled at two different rates: sparse (undersampled) and dense (oversampled).

    The function:
        - Randomly selects one frequency that is resolvable by both samplings,
          and one that is only resolvable by the dense sampling grid (and thus
          aliases under sparse sampling).
        - Generates two synthetic signals using those frequencies.
        - Computes and displays their power spectra using unitary FFT.
        - Highlights the frequency range where aliasing occurs.

    The plot clearly shows how undersampling leads to distortion in the frequency
    domain due to unresolved high-frequency components being aliased.

    Returns:
        None. Displays a matplotlib plot.
    """

    def frequencies_bins_picker(
        sparse_frequencies: np.ndarray, dense_frequencies: np.ndarray
    ) -> np.ndarray:
        """
        Randomly selects two frequency bins:
            - One that exists in both sparse and dense frequency grids (resolvable).
            - One that exists only in the dense grid but not in the sparse grid
            (aliased under sparse sampling).

        Parameters:
            sparse_frequencies (np.ndarray): Frequency bins of the sparse FFT grid.
            dense_frequencies (np.ndarray): Frequency bins of the dense FFT grid.

        Returns:
            np.ndarray: Array containing one resolved and one aliased frequency.
        """
        resolved_frequency = np.random.choice(sparse_frequencies[2:-2], size=1)[-1]
        aliased_frequency = np.random.choice(
            np.setxor1d(sparse_frequencies, dense_frequencies)[2:-2], size=1
        )[-1]
        return np.array([resolved_frequency, aliased_frequency])

    def signal_generator(time: np.ndarray, significant_freq: np.ndarray) -> np.ndarray:
        """
        Generates a complex-valued signal using a full spectrum of frequencies,
        with dominant components at selected "significant" frequencies.

        All other frequencies are populated with a very small amplitude to
        maintain full spectral structure and minimize FFT numerical artifacts.

        Parameters:
            time (np.ndarray): Time vector over [0, 1), unitless and evenly spaced.
            significant_freq (np.ndarray): Array of frequencies with dominant amplitude.

        Returns:
            np.ndarray: Complex-valued synthetic signal.
        """
        N = len(time)
        signal = np.zeros_like(time, dtype=complex)
        for freq in np.fft.fftfreq(N, d=1 / N):
            if freq not in significant_freq:
                magnitude = 1e-6 * np.sqrt(1 / N)
                phase = 0.0
                signal += magnitude * np.exp(1j * (2 * np.pi * freq * time + phase))
        for freq in significant_freq:
            magnitude = np.sqrt(1 / N)
            phase = 0.0
            signal += magnitude * np.exp(1j * (2 * np.pi * freq * time + phase))
        return signal

    sparse_time = np.linspace(0, 1, 20, endpoint=False)
    dense_time = np.linspace(0, 1, 40, endpoint=False)
    sparse_frequencies = np.fft.fftfreq(sparse_time.size, 1 / sparse_time.size)
    dense_frequencies = np.fft.fftfreq(dense_time.size, 1 / dense_time.size)

    significant_frequencies = frequencies_bins_picker(
        sparse_frequencies, dense_frequencies
    )

    sparse_signal = signal_generator(sparse_time, significant_frequencies)
    dense_signal = signal_generator(dense_time, significant_frequencies)

    bar_width = 0.3
    plt.figure(figsize=(16, 9), dpi=160)
    plt.bar(
        np.fft.fftshift(dense_frequencies) - bar_width / 2,
        np.fft.fftshift(np.abs(np.fft.fft(dense_signal, norm="ortho")) ** 2),
        width=bar_width,
        label="oversample",
    )
    plt.bar(
        np.fft.fftshift(sparse_frequencies) + bar_width / 2,
        np.fft.fftshift(np.abs(np.fft.fft(sparse_signal, norm="ortho")) ** 2),
        width=bar_width,
        label="undersample",
    )
    plt.plot(
        [2 * np.min(dense_frequencies), 2 * np.max(dense_frequencies)],
        [1, 1],
        "k--",
    )
    plt.fill_betweenx(
        [0, 2],
        np.min(sparse_frequencies),
        np.max(sparse_frequencies),
        facecolor="red",
        alpha=0.1,
    )
    plt.fill_betweenx(
        [0, 2],
        10 * np.min(sparse_frequencies),
        np.min(sparse_frequencies),
        facecolor="gray",
        edgecolor="gray",
        alpha=0.3,
    )
    plt.fill_betweenx(
        [0, 2],
        np.max(sparse_frequencies),
        10 * np.max(sparse_frequencies),
        facecolor="gray",
        edgecolor="gray",
        alpha=0.3,
    )
    plt.legend(loc=1)

    plt.xticks(dense_frequencies)
    plt.xlabel("Frequency (Hz)")
    plt.xlim(-20, 20)
    plt.ylabel("Power")
    plt.ylim(0, 1.1)

    plt.title(f"Aliasing")
    plt.show()

    return None
```

## **Spectral Leakage**

### Recap

As discussed in Sections **1.2 Windowing in CFT** and **3.2 DTFT of a Windowed, Sampled Signal**, applying a window function to truncate an infinite-length signal—whether in the continuous or discrete domain—introduces convolution in the frequency domain. Specifically, the original spectrum $\hat{f}(\omega)$ is convolved with the Fourier transform of the window function $\hat{w}(\omega)$, which is typically sinc-shaped for a rectangular window.

As a result, **spectral leakage** arises as a fundamental consequence of signal truncation:

$$
\hat{f}_w(\omega) = \hat{f}(\omega) * \hat{w}(\omega)
= \hat{f}(\omega) * \left[L \cdot \mathrm{sinc}(\omega L)\right]
$$

This convolution causes energy from sharp frequency components to "leak" into neighboring frequencies, resulting in a smeared or blurry spectrum.

### Spectral Leakage in DFT and DTFT

Since all signals in practical applications are finite in duration, **spectral leakage is inherent in every DTFT**.

In the case of the DFT, which samples the DTFT at discrete frequency intervals, **spectral leakage is also inherent in every DFT**. 

For ideal signals with frequencies that align perfectly with DFT bins, spectral leakage may appear suppressed—but it's merely hidden, not eliminated. In general, DFT still exhibits leakage due to the finite duration of the input signal.

```python
def display_zero_padding(
    padding_ratio: int = 100, L: float = 1.0, N: int = 20, mode: str = "aligned"
):
    """
    Visualizes spectral leakage in the DFT by comparing sparse and zero-padded frequency spectra.

    This function demonstrates how spectral leakage is present in DFT.
    By comparing a sparse DFT (low frequency resolution) and a zero-padded
    DFT (high frequency resolution), it shows how leakage may be hidden or
    revealed depending on whether the signal frequency **aligns** with DFT bins.

    Parameters:
        padding_ratio (int): Factor by which the signal is zero-padded to improve frequency resolution.
        L (float): Duration of the signal in time.
        N (int): Number of time-domain samples (before zero-padding).
        mode (str): Determines the alignment of the signal frequency with DFT bins.
                    - "aligned": chooses a frequency that aligns with original DFT bins (leakage appears minimal).
                    - "misaligned": chooses a frequency that does not align, making leakage visually apparent.

    Returns:
        None. Displays a plot showing:
            - The sparse DFT (original resolution),
            - The dense DFT (with zero-padding),
            - The theoretical DTFT (sinc² profile).
    """

    def frequencies_bins_picker(frequencies: np.ndarray, range: float = 3 / L) -> float:
        """
        Selects a frequency from the given array, avoiding edges for better plot clarity.

        Parameters:
            frequencies (np.ndarray): Array of candidate frequency bins.
            range (float): Margin to exclude near the minimum and maximum values.
                        Only frequencies strictly within (min + range, max - range) are considered.

        Returns:
            float: A single frequency randomly chosen from the center region of the input array.
        """
        lo, hi = frequencies.min() + range, frequencies.max() - range
        masked_freq = frequencies[(frequencies > lo) & (frequencies < hi)]
        return np.random.choice(masked_freq, size=1, replace=False)

    def signal_generator(time: np.ndarray, significant_freq: np.ndarray) -> np.ndarray:
        """
        Generates a complex sinusoidal signal and its zero-padded version.

        The signal is normalized to maintain consistent energy in plots.
        Padding increases frequency resolution in the DFT.

        Parameters:
            time (np.ndarray): Time vector over which to evaluate the signal.
            significant_freq (np.ndarray): Frequency at which the sinusoid is generated.

        Returns:
            Tuple[np.ndarray, np.ndarray]: A tuple containing:
                - The original complex sinusoid (normalized).
                - The zero-padded version of the signal, also energy-scaled.
        """
        signal = np.exp(1j * 2 * np.pi * significant_freq * time)
        signal /= np.sqrt(np.sum(np.abs(signal) ** 2))

        padded_signal = np.pad(signal, (0, int((padding_ratio - 1) * len(signal))))
        padded_signal *= np.sqrt(padding_ratio)
        return signal, padded_signal

    time = np.linspace(0, L, N, endpoint=False)

    dt = time[1] - time[0]
    frequencies = np.fft.fftfreq(time.size, dt)
    padded_frequencies = np.fft.fftfreq(time.size * padding_ratio, dt)

    if mode == "aligned":
        candidate_freqs = frequencies
    elif mode == "misaligned":
        candidate_freqs = np.setdiff1d(padded_frequencies, frequencies)
    else:
        raise ValueError

    significant_freq = frequencies_bins_picker(candidate_freqs)
    signal, padded_signal = signal_generator(time, significant_freq)

    spectrum = np.abs(np.fft.fft(signal, norm="ortho")) ** 2
    padded_spectrum = np.abs(np.fft.fft(padded_signal, norm="ortho")) ** 2
    theoretical_spectrum = np.sinc(L * (padded_frequencies - significant_freq)) ** 2

    plt.figure(figsize=(16, 9), dpi=160)
    plt.bar(
        frequencies,
        spectrum,
        width=0.03,
        color="tab:red",
        label="sparse DFT",
        zorder=2,
    )

    plt.bar(
        padded_frequencies,
        padded_spectrum,
        width=0.8 / padding_ratio / L,
        color="tab:purple",
        label="dense DFT",
        alpha=0.5,
        zorder=-1,
    )
    plt.plot(
        np.fft.fftshift(padded_frequencies),
        np.fft.fftshift(theoretical_spectrum),
        "k--",
        zorder=5,
        label="theoretical DTFT",
    )

    plt.legend(loc=1)
    plt.xlim(significant_freq - 3 / L, significant_freq + 3 / L)
    plt.ylim(0, 1.1)
    plt.xlabel("$\omega$")
    plt.ylabel("Power Spectral Density")
    plt.title("DFT might make spectral leakage invisible")

    plt.show()

    return None
```

## **Convolution**

### Linear Convolution

#### Infinite-Length Signals

The **linear convolution** of two signals $p[n]$ and $q[n]$ is defined as:
$$
y[n] = (p * q)[n] = \sum_{m=-\infty}^{\infty} p[m] \cdot q[n - m], \quad n \in \mathbb{Z}
$$

This definition assumes that both $p[n]$ and $q[n]$ are defined over the entire set of integers $\mathbb{Z}$, and the sum is well-defined when at least one of the signals is absolutely summable.

#### Finite-Length Signals

In practical scenarios, signals are typically of **finite duration**. To apply the convolution formula in such cases, we extend both signals by **zero-padding**, i.e., we define them to be zero outside their original support. This allows us to retain the standard convolution expression while ensuring that the summation remains finite and computationally feasible.

Suppose:
- $p[n]$ is supported on $\text{supp}(p) = \{0, 1, \dots, N - 1\}$
- $q[n]$ is supported on $\text{supp}(q) = \{0, 1, \dots, M - 1\}$

Then the linear convolution simplifies to:

$$
y[n] = (p * q)[n] = \sum_{m=0}^{N-1} p[m] \cdot q[n - m]
$$

---

### Circular Convolution

In contrast to linear convolution, which assumes zero-padding beyond signal support, **circular convolution** operates under the assumption that one or both sequences are **periodic**.

Assume $q[n]$ is periodic with period $N$. Let its periodic extension be denoted as $q_N[n]$, satisfying:

$$
q_N[n] = q_N[n + N]
$$

Then, the circular convolution of $p[n]$ and $q_N[n]$ is initially defined as:

$$
y[n] = (p \circledast q)[n] = \sum_{m=0}^{N-1} \left( \sum_{k=-\infty}^{\infty} p[m + kN] \right) \cdot q_N[n - m]
$$

Although mathematically precise, this expression is often more general than needed. In most practical applications, both signals are assumed to be **finite-length and periodic** with period $N$. Assume:

- Both $p[n]$ and $q_N[n]$ are defined over the interval $[0, N - 1]$
- Both sequences are extended periodically, i.e., $p[n + kN] = p[n]$

Under these assumptions, the inner infinite sum collapses to $p[m]$, resulting in the simplified and widely-used form:

$$
y[n] = (p \circledast q_N)[n] = \sum_{m=0}^{N-1} p[m] \cdot q_N[(n - m) \bmod N]
$$

Here, the modulo operation introduces the **wrap-around behavior** characteristic of circular convolution. The output $y[n]$ is also periodic with period $N$, satisfying $y[n+kN] = y[n]$.

---

### Key Differences: Linear vs. Circular Convolution

At a glance, the mathematical forms of linear and circular convolution may appear similar. However, they differ fundamentally in assumptions and outcomes:

- **Linear convolution** treats signals as zero beyond their support. The result captures the complete overlap behavior and has a length of $N + M - 1$, where $N$ and $M$ are the lengths of the input sequences.
- **Circular convolution** assumes periodic signals. It wraps indices modulo $N$, producing a periodic result of length $N$.

---

### Practical Consideration: Matching Signal Lengths

To compute circular convolution, both input signals must be of **equal length**. If one signal is shorter—say, $p[n]$ has length $N$, and $q[n]$ has length $M > N$—we pad the shorter sequence with zeros to match the longer one.

For example, if $p[n]$ is shorter, its padded version becomes:

$$
p_{\text{padded}}[n] = 
\begin{cases}
p[n], & 0 \leq n < N \\
0, & N \leq n < M 
\end{cases}
$$

This zero-padding ensures both sequences are periodic over a common length $M$, enabling consistent circular convolution.

---

## **Convolution in the DFT**

### Recap

#### Continuous Fourier Transform (CFT)

The **convolution-multiplication duality** in the CFT domain is a foundational result:

In the CFT, a fundamental property is the convolution-multiplication duality, which states:

$$
p(t) \cdot q(t) \longleftrightarrow P(f) * Q(f)
$$

$$
p(t) * q(t) \longleftrightarrow P(f) \cdot Q(f)
$$

These relationships show that multiplication in one domain corresponds to convolution in the other, and vice versa.

#### Discrete Fourier Transform (DFT)

This duality does **not** carry over directly to the DFT. The reason lies in the **periodicity** inherent in the DFT, which treats signals as periodic in both time and frequency domains.

As a result, the DFT interprets convolution as **circular convolution**, due to this assumed periodicity. The modified duality becomes:

$$
p[n] \cdot q[n] \xleftrightarrow{\text{DFT}} P[k] \circledast Q[k]
$$

$$
p[n] \circledast q[n] \xleftrightarrow{\text{DFT}} P[k] \cdot Q[k]
$$

Here:
- $\circledast$ denotes circular convolution.
- Pointwise multiplication in one domain corresponds to circular convolution in the other.

Thus, while the CFT maintains a clean convolution-multiplication duality, the DFT adapts this relationship for a periodic signal model.

---

### Computing Circular Convolution via DFT

The DFT provides an efficient method to compute circular convolution. The procedure is as follows:

1. **Zero-pad** both signals (if necessary) to equal length $M$
2. Compute the **DFT** of each signal
3. Perform **pointwise multiplication** of the DFTs
4. Apply the **inverse DFT** to obtain the result

Formally:

$$
y[n] = (p \circledast q)[n] = \mathrm{iDFT}\left( \mathrm{DFT}(p[n]) \cdot \mathrm{DFT}(q[n]) \right)
$$

---

### Computing Linear Convolution via DFT

Although the DFT naturally yields circular convolution, we can compute linear convolution using the DFT by padding the input sequences appropriately:

1. Zero-pad both signals to a length of at least $N + M - 1$
2. Compute the **DFT** of the zero-padded signals
3. Multiply the DFTs **pointwise**
4. Apply the **inverse DFT** to recover the result

With sufficient zero-padding, wrap-around effects are eliminated, and the output matches the linear convolution:

$$
y[n] = (p * q)[n] = \mathrm{iDFT}\left( \mathrm{DFT}(p_{\text{padded}}[n]) \cdot \mathrm{DFT}(q_{\text{padded}}[n]) \right)
$$

```python
def display_convolution(length: int = 9):
    """
    Demonstrates and visualizes linear and circular convolution using both using primitive
    and FFT-based methods.

    The resulting signals are plotted for comparison, with clear visual alignment and
    boundary effects marked.

    Parameters:
        length (int, optional): The half-width of the averaging kernel used for convolution.
                                The full kernel has size 2 * length + 1. Default is 9.

    Notes:
        The plot aligns all convolution results by shifting them so that the center of the
        kernel corresponds to the input signal indices. For both linear and circular
        convolution, the output is shifted by floor(min(M, N) / 2), where M and N are the
        lengths of the signal and kernel.

        Linear convolution is computed in 'full' mode, which includes partially overlapped
        regions at the beginning and end. To produce a centered result suitable for
        visualization, a slice of length max(M, N) is extracted after shifting. This ensures
        the output is aligned with the input while preserving some partial contributions near
        the boundaries—effectively mimicking 'same' mode but with explicit alignment control.

        Circular convolution is inherently periodic and produces output of length max(M, N).
        The same alignment shift is applied to ensure consistent comparison with the linear
        results.

        Gray shading in the plot marks regions where boundary effects (due to zero-padding or
        circular wrap-around) influence the result.

    Returns:
        None
    """

    def linear_convolution(
        signal_1: np.ndarray, signal_2: np.ndarray, method: str, mode: str
    ) -> np.ndarray:
        """
        Computes the linear convolution of two 1D signals using the specified method and cropping mode.

        This function supports both a direct time-domain implementation ("primitive") and a
        frequency-domain implementation using the Fast Fourier Transform ("fft"). The result
        can be cropped to match one of three standard modes: 'full', 'same', or 'valid'.

        Parameters:
            signal_1 (np.ndarray): The first 1D input signal (p).
            signal_2 (np.ndarray): The second 1D input signal (q).
            method (str): Method to use for convolution. Must be either:
                        - "primitive": direct time-domain implementation
                        - "fft": convolution via the FFT
            mode (str): Output shape mode. Must be one of:
                        - "full": returns the full linear convolution (length N + M - 1)
                        - "same": returns the central part of the convolution (length ≈ max(N, M))
                        - "valid": returns only parts where signals fully overlap (length = |N - M| + 1)

        Returns:
            np.ndarray: The final convolved result cropped according to the selected mode.

        Raises:
            ValueError: If the method or mode is not one of the accepted options.
        """

        def primitive(signal_1: np.ndarray, signal_2: np.ndarray):
            """
            Computes the **linear convolution** of two 1D signals using a manual time-domain approach.

            This function pads the input signals to the required length, manually flips and
            shifts one of the signals, and performs element-wise multiplication followed by
            summation to compute the convolution.

            Parameters:
                signal_1 (np.ndarray): The first 1D input signal.
                signal_2 (np.ndarray): The second 1D input signal.

            Returns:
                np.ndarray: The convolved signal as a 1D array.
            """
            total_length = signal_1.size + signal_2.size - 1

            padded_signal_1 = np.pad(
                signal_1, pad_width=(0, total_length - signal_1.size)
            )
            padded_signal_2 = np.pad(
                signal_2, pad_width=(0, total_length - signal_2.size)
            )

            result = np.zeros_like(padded_signal_1)

            for n in range(result.size):
                p_m = padded_signal_2
                q_nm = np.roll(np.flip(padded_signal_1), n + 1)
                result[n] = np.sum(p_m * q_nm)

            return result

        def fft(signal_1: np.ndarray, signal_2: np.ndarray):
            """
            Computes the **linear convolution** of two 1D signals using the Fast Fourier Transform (FFT).

            This method takes advantage of the convolution theorem, which states that
            convolution in the time domain is equivalent to pointwise multiplication
            in the frequency domain.

            Parameters:
                signal_1 (np.ndarray): The first 1D input signal.
                signal_2 (np.ndarray): The second 1D input signal.

            Returns:
                np.ndarray: The convolved signal as a 1D array.
            """
            total_length = signal_1.size + signal_2.size - 1

            padded_signal_1 = np.pad(
                signal_1, pad_width=(0, total_length - signal_1.size)
            )
            padded_signal_2 = np.pad(
                signal_2, pad_width=(0, total_length - signal_2.size)
            )

            result = np.fft.ifft(
                np.fft.fft(padded_signal_1) * np.fft.fft(padded_signal_2)
            ).real

            return result

        def crop_output(full_result: np.ndarray, len_p: int, len_q: int):
            """
            Crops the full convolution result based on the specified mode.

            Parameters:
                full_result (np.ndarray): The full result of combining signals p and q,
                                        of length len_p + len_q - 1.
                len_p (int): Length of signal p.
                len_q (int): Length of signal q.

            Returns:
                np.ndarray: Cropped result according to the selected mode: 'valid', 'same', or 'full'.
            """
            if mode == "full":
                return full_result
            elif mode == "same":
                min_len = min(len_p, len_q)
                start = int(
                    np.floor((min_len - 1) / 2)
                )  # first floor(min_len/2) elements
                end = -int(np.ceil((min_len - 1) / 2))  # last ceil(min_len/2) elements
                return full_result[start:end]
            elif mode == "valid":
                min_len = min(len_p, len_q)
                start = min_len - 1  # first min_len elements
                end = -min_len + 1  # last min_len elements
                return full_result[start:end]
            else:
                raise ValueError(
                    f"Invalid mode '{mode}'. Use 'valid', 'same', or 'full'."
                )

        if method == "primitive":
            convolved = primitive(signal_1, signal_2)
        elif method == "fft":
            convolved = fft(signal_1, signal_2)
        else:
            raise ValueError(f"Invalid method '{method}'. Use 'primitive' or 'fft'.")

        final_result = crop_output(convolved, len_p=signal_1.size, len_q=signal_2.size)

        return final_result

    def circular_convolution(signal_1: np.ndarray, signal_2: np.ndarray, method: str):
        def primitive(signal: np.ndarray, kernel: np.ndarray):
            """
            Computes the **circular convolution** of two 1D signals using a direct time-domain method.

            This method flips and circularly shifts the input signal, then computes the element-wise
            product with the (zero-padded) kernel and sums the result at each position.

            Parameters:
                signal (np.ndarray): The input 1D signal.
                kernel (np.ndarray): The second 1D signal to convolve with (the kernel). It is
                                    zero-padded to match the length of the signal if shorter.

            Returns:
                np.ndarray: The circular convolution result, with the same length as `signal`.
            """
            padded_signal = signal
            padded_kernel = np.pad(
                array=kernel, pad_width=((0, signal.size - kernel.size),)
            )
            result = np.zeros_like(padded_signal)
            for n in range(result.size):
                p_m = padded_kernel
                q_nm = np.roll(np.flip(padded_signal), n + 1)
                result[n] = np.sum(p_m * q_nm)
            return result

        def fft(signal: np.ndarray, kernel: np.ndarray):
            """
            Computes the **circular convolution** of two 1D signals using the Fast Fourier Transform (FFT).

            Circular convolution in the time domain is equivalent to pointwise multiplication
            in the frequency domain. This method zero-pads the shorter signal to match the length
            of the longer one before applying the FFT-based multiplication.

            Parameters:
                signal (np.ndarray): The input 1D signal.
                kernel (np.ndarray): The second 1D signal to convolve with. It is zero-padded
                                    to match the length of the signal if shorter.

            Returns:
                np.ndarray: The circular convolution result as a complex-valued array.
                            Use `.real` to extract real-valued output if inputs are real.
            """
            padded_signal = signal
            padded_kernel = np.pad(
                array=kernel, pad_width=((0, signal.size - kernel.size),)
            )
            result = np.fft.ifft(
                np.fft.fft(padded_signal) * np.fft.fft(padded_kernel)
            ).real
            return result

        if len(signal_2) > len(signal_1):
            signal_1, signal_2 = signal_2, signal_1
        if method == "primitive":
            signal = primitive(signal=signal_1, kernel=signal_2)
        elif method == "fft":
            signal = fft(signal=signal_1, kernel=signal_2)
        return signal

    def convolutional_kernel(length) -> np.ndarray:
        """
        Creates a moving averaging kernel.

        Parameters:
            length (int): The half-width of the kernel. The full kernel will have size 2*length + 1.

        Returns:
            np.ndarray: A 1D normalized box filter of length (2 * length + 1).
        """
        kernel = np.ones(shape=(2 * length + 1,))
        kernel /= kernel.size
        return kernel

    def signal_generator(time: np.ndarray) -> np.ndarray:
        """
        Generates a smooth 1D signal.

        Parameters:
            time (np.ndarray): A 1D array of time values, which also defines the output signal length.

        Returns:
            np.ndarray: A smoothed 1D signal of the same length as `time`.
        """
        signal = np.random.normal(10.0, 3, size=(time.size,))
        kernel = convolutional_kernel(1)
        for _ in range(10):
            signal = np.roll(
                circular_convolution(signal, kernel, method="fft").real, -1
            )
        return signal

    time = np.linspace(0, 1, 100)
    kernel = convolutional_kernel(length=length)
    signal = signal_generator(time)

    # --- Constant ---
    shift = int(np.floor(min(signal.size, kernel.size) / 2))
    output_length = max(signal.size, kernel.size)

    # --- Linear Convolution Results ---
    linear_convolution_numpy_full = np.convolve(signal, kernel, mode="full")
    linear_convolution_numpy = np.roll(linear_convolution_numpy_full, -shift)[
        :output_length
    ]

    linear_convolution_primitive_full = linear_convolution(
        signal, kernel, method="primitive", mode="full"
    )
    linear_convolution_primitive = np.roll(linear_convolution_primitive_full, -shift)[
        :output_length
    ]

    linear_convolution_fft_full = linear_convolution(
        signal, kernel, method="fft", mode="full"
    )
    linear_convolution_fft = np.roll(linear_convolution_fft_full, -shift)[
        :output_length
    ]

    # --- Circular Convolution Results ---
    circular_convolution_primitive = circular_convolution(
        signal, kernel, method="primitive"
    )
    circular_convolution_primitive = np.roll(circular_convolution_primitive, -shift)

    circular_convolution_fft = circular_convolution(signal, kernel, method="fft")
    circular_convolution_fft = np.roll(circular_convolution_fft, -shift)

    plt.figure(figsize=(16, 9), dpi=160)
    plt.plot(time, signal, "k-", lw=4, label="original signal", alpha=0.8)
    plt.plot(
        time,
        linear_convolution_numpy,
        "b-.",
        lw=2,
        label="linear convolution (np.convolve)",
    )
    plt.plot(
        time,
        linear_convolution_primitive,
        "r--",
        lw=2,
        label="linear convolution (primitive)",
    )
    plt.plot(
        time,
        linear_convolution_fft,
        "g:",
        lw=2,
        label="linear convolution (fft)",
    )
    plt.plot(
        time,
        circular_convolution_primitive,
        c="tab:red",
        ls="--",
        lw=7,
        label="circular convolution (primitive)",
        alpha=0.4,
    )
    plt.plot(
        time,
        circular_convolution_fft,
        c="tab:blue",
        ls="-.",
        lw=7,
        label="circular convolution (fft)",
        alpha=0.4,
    )
    plt.fill_betweenx(
        [-20, 20],
        time[0],
        time[0 + length],
        facecolor="gray",
        edgecolor="white",
        alpha=0.25,
    )
    plt.fill_betweenx(
        [-20, 20],
        time[-1 - length],
        time[-1],
        facecolor="gray",
        edgecolor="white",
        alpha=0.25,
    )
    plt.legend(loc=1)
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    plt.xlim(0, 1)
    plt.ylim(5.0, 15.0)
    plt.title(
        "Linear Convolution and Circular Convolution\nGray shading depict B.C. effect area"
    )
    plt.show()

    return None
```
